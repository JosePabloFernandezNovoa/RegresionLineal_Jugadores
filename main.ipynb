```python
import numpy as np 
import pandas as pd 

from sklearn import linear_model #modulo que implementa la clase LinearRegresion
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score


from sklearn.model_selection import train_test_split

jugadores=pd.read_csv("./all_players.csv")

tiro=jugadores["Shooting"]
media=jugadores["Overall"]

media_train, media_test, tiro_train, tiro_test=train_test_split(tiro, media, test_size=0.2)

plt.scatter(media_train, tiro_train, label='Datos de entrenamiento', color='r')
plt.scatter(media_test, tiro_test, label='Datos de prueba', color='g')

plt.legend()
plt.title("Test train split")
plt.show()
```

```python

regr = LinearRegression()

media_train=np.array(media).reshape(-1, 1)
tiro_train=np.array(tiro).reshape(-1, 1)

regr.fit(media_train, tiro_train)

```

```python
print('Coeficiente: \n', regr.coef_)
# Este es el valor donde corta el eje Y (en X=0)
print('Trmino independiente: \n', regr.intercept_)
```

```python
pred = regr.predict(media_train)

print("Error Cuadrtico medio: %.2f" % mean_squared_error(media_train, pred))
print('Coeficiente de determinacin: %.2f' % r2_score(tiro_train, pred))
```
